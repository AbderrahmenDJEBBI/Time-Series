# Exemple 1 : Air Passengers (univarié)
Les données traduisent l'évolution du nombre mensuel de passagers aériens américains sur des vols internationaux en milliers de 1949 à 1960.

```python
data(AirPassengers)
AP <-AirPassengers
```

```python
length(which(is.na(AP)==TRUE))
```

```python
plot(AP,xlab="Time", ylab="Number of AirPassengers (1000)")
```

```python
AP.year <- aggregate(AP, FUN="mean")
plot(AP.year,type="o")
summary(AP)
```

```python
#boxplot(AP)
#show(cycle(AP))
boxplot(AP~cycle(AP))
```

```python
lag.plot(AP,lag=12,do.lines=F)
```

```python
acf(AP,lag=36)
```

# Exemple 2 : Chocolat - Bière - Electricity (multivarié)
Les données traduisent l'approvisionnement mensuel de l'électricité (millions de kWh), bière (Ml),
et  chocolat (tonnes) en  Australie pour la période du Janvier 1958 jusqu'au Décembre 1990. 

```python
CBE <- read.table("cbe.dat",header=TRUE)
```

```python
class(CBE)
```

```python
Choc.ts<-ts(CBE[,1],start=c(1958,1),freq=12)
Beer.ts<-ts(CBE[,2],start=c(1958,1),freq=12)
Elec.ts<-ts(CBE[,3],start=c(1958,1),freq=12)
```

```python
plot(cbind(Choc.ts,Beer.ts,Elec.ts),xlab="Time")
```
